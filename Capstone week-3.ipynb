```python
table =data[0]
```

###### The dataframe consists of three columns: PostalCode, Borough, and Neighborhood

```python
table.columns
```

###### Ignore cells with a borough that is Not assigned

```python
table['Borough'].replace("Not assigned",np.nan,inplace=True)
```

```python
table.head()
```

```python
table.dropna(inplace=True)
```

```python
table.head()
```

```python
table =table.groupby(['Postal Code','Borough'])['Neighborhood'].apply(', '.join).reset_index()
table.head()
```

###### If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
table["Neighborhood"].replace("Not assigned",table["Borough"],inplace = True)
```

```python
import folium
```

```python
df3 = coord_table.copy()
df3 = df3[coord_table.Borough.str.contains("Toronto")]
df3.head()
```

```python
toronto_map = folium.Map(location=[43.65, -79.4], zoom_start=12)

X = df3['Latitude']
Y = df3['Longitude']
Z = np.stack((X,Y),axis=1)
```

```python
kmeans =KMeans(n_clusters=4, random_state=0).fit(Z)
```

```python
clusters=kmeans.labels_
colors = ['red','green','blue','yellow']
df3['Cluster']=clusters
```

```python
for latitude, longitude, borough, cluster in zip(df3['Latitude'], df3['Longitude'], df3['Borough'], df3['Cluster']):
    label = folium.Popup(borough, parse_html=True)
    folium.CircleMarker([latitude, longitude],
                      radius=5, popup=label,
                      color = 'black',
                      fill=True,
                      fill_color=colors[cluster],
                      fill_opacity=0.7).add_to(toronto_map)
toronto_map
```
